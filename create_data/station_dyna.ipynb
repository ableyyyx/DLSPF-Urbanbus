```python
import numpy as np
# 构建公交站点的邻接矩阵无向图 temp_mat[i][j] = 1 表示站点 i 和 j 在某条公交线路上是相邻站点

# 创建一个方阵 temp_mat，其形状是 (站点数, 站点数)，所有元素初始化为 0
temp_mat=np.zeros((len(unique_flattened_list),len(unique_flattened_list)),dtype=int)

for bus_dir in [phy_graph_0,phy_graph_1]:
    for bus_keys in bus_dir.keys():
        # temp_data 是当前公交车停靠的所有站点 ID 的列表，按顺序排列
        temp_data=list(bus_dir[bus_keys].Stop_stn)
        # 使用 zip() 同时遍历 temp_data 里的站点 ID（stn_value）和索引 i，i 代表站点在 temp_data 里的顺序（即公交到达的顺序）
        for (stn_value,i) in zip(temp_data,range(0,len(temp_data))):
            # 处理路线起点
            if i==0:
                # 让第一个站点 temp_data[i] 连接到下一个站点 temp_data[i+1]
                temp_mat[temp_dict[temp_data[i]]][temp_dict[temp_data[i+1]]]=1
            # 处理路线终点
            elif i==(len(temp_data)-1):
                # 让最后一个站点 temp_data[i] 连接到前一个站点 temp_data[i-1]
                temp_mat[temp_dict[temp_data[i]]][temp_dict[temp_data[i-1]]]=1
            # 处理中间站点
            else:
                print(stn_value)
                temp_mat[temp_dict[temp_data[i]]][temp_dict[temp_data[i+1]]]=1
                temp_mat[temp_dict[temp_data[i]]][temp_dict[temp_data[i-1]]]=1
```

```python
# for i in range(0,3):

def find_max_index(phy_graphy):
    a=[]
    for i in range(0,phy_graphy.shape[0]):
        a.append(sum(phy_graphy[i,:]))
    if max(a)==1:
        return False
    # 返回：
    # 1.a —— 每个站点的度数列表。
    # 2.a.index(max(a)) —— 连接最多站点的站点索引 
    # 3.np.where(phy_graphy[a.index(max(a)), :] == 1)[0].tolist() —— 该站点连接的所有站点索引
    return a,a.index(max(a)),np.where(phy_graphy[a.index(max(a)),:]==1)[0].tolist()

print(find_max_index(phy_graphy))
```
